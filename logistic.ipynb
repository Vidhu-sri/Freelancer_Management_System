```python
import pandas as pd
import numpy as np
import statistics as stat
import math
```

```python
df=pd.read_csv("C:\\Users\\Lenovo\\Desktop\\manas stuff\\train (1).csv")
df.drop(['Name'], axis=1, inplace=True)  
df
```

```python
x_trans=[]
for i in range(1,7):
    x_trans.append(df.iloc[:, i])
x_trans=np.array(x_trans, dtype=float)
#z-score normalization

for i in range(6):
    mean=stat.mean(x_trans[i])
    std=stat.stdev(x_trans[i])
    for j in range(577):
            nr=x_trans[i][j]-mean
            dr=std
            x_trans[i][j]=nr/dr
x=x_trans.T

```

```python
class logistic_regression:
    def __init__(self):
        self.columns=5
        self.rows=577
        self.alpha=0.01
        self.w=np.zeros(6)
        self.b=0
        self.iterations=1000
    
    def g(self, i):
        return (np.dot(self.w, x[i])+self.b)
    
    def S(self, i):    #sigmoid function
        nr=1
        dr=1+math.exp(-self.g(i))
        return nr/dr
    
    def p(self, i):
        y=df["Survived"][i]
        p1=self.S(i)**y
        p2=(1-self.S(i))**(1-y)
        return p1*p2
    
    def J(self):  #cost function
        l=[]
        for i in range(self.rows):
            y=df["Survived"][i]
            log_1=y*math.log(self.S(i))
            log_2=(1-y)*math.log((1-self.S(i)))
            l.append(log_1 * log_2)
        l=np.array(l, dtype=float)
        return -np.sum(l)/(self.rows)
    
    
    #gradient descent
    def iterate(self):
        pred=[]
        for i in range(self.rows):
            pred.append(self.S(i))
        pred=np.array(pred, dtype=float)
        actual=np.array(df["Survived"], dtype=float)
        diff=np.subtract(pred, actual)
        self.w-= (self.alpha/self.rows)*np.dot(x_trans, diff.T)
        self.b-= (self.alpha/self.rows)*np.sum(diff)
        self.J()




        
```

```python
j=[]
obj = logistic_regression()
for i in range(obj.iterations):
    obj.iterate()
    j.append(obj.J())
print(obj.w, '\t', obj.b)
```

```python
for i in range(obj.rows):
    if obj.S(i)>=0.5:
        print(1, "\t", obj.S(i), "\t", df["Survived"][i])
    else:
        print(0, "\t", obj.S(i), "\t", df["Survived"][i])
```
